```python
tips = sns.load_dataset('tips')
tips.head()
```

- 식사인원이 5명 이상인 저녁 테이블
- 흡연자가 없는 주말 테이블
- total_bill, tip 출력

```python
c1 = tips['size'] >= 5
c2 = tips['time'] == 'Dinner'
c3 = tips['day'].isin(['Sat', 'Sun'])
c4 = tips['smoker'] == 'No'

tips.loc[(c1 & c2) & (c3 & c4), ['total_bill', 'tip']]
```
