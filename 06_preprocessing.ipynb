### to_numeric

```python
df1.info()
```

```python
df1.head()
```

```python
df1['운동량'] = pd.to_numeric(df1['운동량'], errors='coerce')
```

```python
df1.info()
```

```python
df1.loc[2344]
```

```python
df1.tail()
```

```python
#kawaii하게 잘라봐야지
bins = [0, 1000, 2000, 3000, df1['운동량'].max()]
pd.cut(df1['운동량'], bins).value_counts()
```

```python
pd.cut(df1['운동량'], bins, labels=['운동부족', '보통', '많음', '운동과다'])
```

```python
df1.describe()
```

```python
df1['운동량_cut'] = pd.cut(df1['운동량'], bins=10)
df1['운동량_cut'].value_counts()
```

*이상치*를 어떻게 처리할것인가?
1. 값을 기준으로 나눔 > 걍cut()
2. 범위를 기준으로 나눔 > qcut()

```python
df1['운동량_qcut'] = pd.qcut(df1['운동량'], q=10)
df1['운동량_qcut'].value_counts()
```

```python
##kawaii하게 자르기 2
qcut_bins = [0, 0.2, 0.8, 1]
pd.qcut(df1['운동량'], qcut_bins)
# 소괄호/초과 미만, 대괄호/이상,이하
```

```python
pd.qcut(df1['운동량'], qcut_bins, labels=['소', '중', '대'])
```
