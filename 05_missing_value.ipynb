## fillna

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
df2 = df.copy()
```

```python
df1['age'].fillna(100) #데이터 프레임에 적용은되지 않음 ,결과값만 보임
df1['age'] = df1['age'].fillna(100) #변수 재할당 후 직접적 수정됨
```

```python
df1['age'].tail()
```

```python
df1['deck'].fillna('C')
```

```python
# df1['deck'].fillna('No Data')
# 카테고리타입에 없는 정보는 fillna 할 수 없음
```

```python
#범주를 추가해서 fillna
df1 = df1['deck'].cat.add_categories('No Data')
df1.fillna('No Data')
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

- 남자 승객의 결측치에는 남자 승객의 평균을 넣고
- 여자 승객의 결측치에는 여자 승객의 평균ㅇ르 넣어라 (나이)
    - 재할당을 못하겠어ㅠㅠ 

```python
c1 = df1['sex'] == 'male'
mean_age_m = df1.loc[c1, 'age'].mean()
df1.loc[c1, 'age'] = df1.loc[c1, 'age'].fillna(mean_age_m)
```

```python
c2 = df1['sex'] == 'female'
mean_age_f = df1.loc[c2, 'age'].mean()
df1.loc[c2, 'age'] = df1.loc[c2, 'age'].fillna(mean_age_f)
```

```python
df1['age'].mean()
```

```python
df1['age'].mean() #재할당 못했을때
```

```python
df1['deck'].mode()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0]).head()
```
